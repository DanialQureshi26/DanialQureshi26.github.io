```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assumption: 1000 test images = 500 Legal + 500 No Ball
total_legal = 500
total_noball = 500

models = [
    ("Stacking (Logistic Regression)", 0.8606, 0.7900),
    ("Weighted Ensemble", 0.8242, 0.8160),
    ("Ensemble 2 (Mob+CNN)", 0.8319, 0.7920),
    ("Stacking (SVM)", 0.8786, 0.7240),
    ("Ensemble 1 (VGG+ResNet)", 0.7478, 0.8540),
    ("VGG16", 0.7431, 0.8560),
    ("MobileNetV2", 0.7689, 0.7320),
    ("Simple CNN", 0.7154, 0.7440),
    ("ResNet50", 0.0000, 0.0000)
]

subplot_labels = ["(a)", "(b)", "(c)", "(d)", "(e)", "(f)", "(g)", "(h)", "(i)"]

class_labels = ["Legal", "No Ball"]

cmaps = [
    "RdYlBu", "RdYlGn", "Purples",
    "coolwarm", "YlOrRd", "Greens",
    "Blues", "Oranges", "Reds"
]

confusion_matrices = []

for model_name, precision, recall in models:

    # Positive class = No Ball
    TP = round(recall * total_noball)

    if precision == 0:
        FP = 0
    else:
        FP = round((TP / precision) - TP)

    FN = total_noball - TP
    TN = total_legal - FP

    cm = np.array([
        [TN, FP],
        [FN, TP]
    ])

    confusion_matrices.append(cm)

# Plot all confusion matrices
fig, axes = plt.subplots(3, 3, figsize=(18, 15))
axes = axes.flatten()

for i, ax in enumerate(axes):
    sns.heatmap(
        confusion_matrices[i],
        annot=True,
        fmt="d",
        cmap=cmaps[i],
        cbar=False,
        xticklabels=class_labels,
        yticklabels=class_labels,
        linewidths=1,
        linecolor="white",
        annot_kws={"size": 13, "weight": "bold"},
        ax=ax
    )

    ax.set_title(
        f"{subplot_labels[i]} {models[i][0]}",
        fontsize=11,
        fontweight="bold",
        pad=10
    )

    ax.set_xlabel("Predicted Label", fontsize=10)
    ax.set_ylabel("True Label", fontsize=10)

plt.tight_layout()

plt.savefig(
    "confusion_matrices_1000_augmented_test_images.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import auc

# -------------------------------
# Model Metrics (YOUR DATA)
# -------------------------------

models = [
    ("Stacking Ensemble (Meta-LR)", 0.8333, 0.9091),
    ("Ensemble 2 (Mob+CNN)", 0.7692, 0.9091),
    ("MobileNetV2", 0.7692, 0.9091),
    ("Weighted Ensemble", 0.7143, 0.9091),
    ("Ensemble 1 (VGG+ResNet)", 0.7143, 0.9091),
    ("MHA Ensemble", 0.7143, 0.9091),
    ("VGG16", 0.7143, 0.9091),
    ("Simple CNN", 0.7273, 0.7273),
    ("ResNet50", 0.0000, 0.0000)
]

# Assume balanced dataset
P = 50   # No Ball
N = 50   # Legal

plt.figure(figsize=(10, 8))

for name, precision, recall in models:

    TPR = recall

    if precision == 0:
        FPR = 0
    else:
        TP = recall * P
        FP = (TP / precision) - TP
        FPR = FP / N

    # Create smooth ROC curve (simple interpolation)
    fpr_points = np.array([0, FPR, 1])
    tpr_points = np.array([0, TPR, 1])

    auc_score = auc(fpr_points, tpr_points)

    plt.plot(fpr_points, tpr_points, marker='o',
             label=f"{name} (AUC={auc_score:.3f})")

# Random baseline
plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')

plt.xlabel("False Positive Rate", fontsize=12)
plt.ylabel("True Positive Rate", fontsize=12)
plt.title("ROC Curves for All Models (Approximate)", fontsize=14, fontweight='bold')
plt.legend(fontsize=9)
plt.grid(alpha=0.3)

plt.tight_layout()
plt.savefig("roc_curves_all_models.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# -------------------------------
# Results Data
# -------------------------------

data = {
    "Model": [
        "Stacking Ensemble (Meta-LR)",
        "Ensemble 2 (Mob+CNN)",
        "MobileNetV2",
        "Weighted Ensemble",
        "Ensemble 1 (VGG+ResNet)",
        "MHA Ensemble",
        "VGG16",
        "Simple CNN",
        "ResNet50"
    ],
    "Accuracy": [86.96, 82.61, 82.61, 78.26, 78.26, 78.26, 78.26, 73.91, 52.17],
    "Precision": [83.33, 76.92, 76.92, 71.43, 71.43, 71.43, 71.43, 72.73, 0.00],
    "Recall": [90.91, 90.91, 90.91, 90.91, 90.91, 90.91, 90.91, 72.73, 0.00]
}

df = pd.DataFrame(data)

# Optional F1-score calculation
df["F1-score"] = 2 * (df["Precision"] * df["Recall"]) / (df["Precision"] + df["Recall"])
df["F1-score"] = df["F1-score"].fillna(0)

# -------------------------------
# Grouped Bar Chart
# -------------------------------

models = df["Model"]
x = np.arange(len(models))
width = 0.2

plt.figure(figsize=(16, 7))

plt.bar(x - 1.5*width, df["Accuracy"], width, label="Accuracy")
plt.bar(x - 0.5*width, df["Precision"], width, label="Precision")
plt.bar(x + 0.5*width, df["Recall"], width, label="Recall")
plt.bar(x + 1.5*width, df["F1-score"], width, label="F1-score")

plt.xlabel("Models", fontsize=12, fontweight="bold")
plt.ylabel("Performance (%)", fontsize=12, fontweight="bold")
plt.title("Performance Comparison of Models", fontsize=14, fontweight="bold")

plt.xticks(x, models, rotation=35, ha="right")
plt.ylim(0, 100)
plt.legend()
plt.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.savefig("model_performance_comparison_bar_chart.png", dpi=300, bbox_inches="tight")
plt.show()
```
